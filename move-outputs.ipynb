```python
# Move outputs to a different directory than the input one.
import glob

json_files = glob.glob("../input/pkna/*/*/*.json")
json_files[:10]
```

```python
# The existing structure is:
# ../input/pkna/pkna-2/gemini-2.0-flash-xyz/out-3.part.json
#
# The new structure will be:
# ../output/transcript/gemini-2.0-flash-xyz/pkna-2/out-3.part.json
import re
import os

pattern = re.compile(r"../input/pkna/(.*)/(.*)/(.*).json")

for json_file in json_files:
    match = pattern.match(json_file)
    if not match:
        print(f"Pattern not matched for {json_file}")
        continue
    # Extract the groups from the match
    pknum = match.group(1)
    model = match.group(2)
    out_file = match.group(3)
    # Move the file
    new_path = f"../output/transcript/{model}/{pknum}/{out_file}.json"
    os.makedirs(os.path.dirname(new_path), exist_ok=True)
    os.rename(json_file, new_path)
```

```python
# Remove empty directories
input_dir = "../input/pkna"

for root, dirs, files in os.walk(input_dir, topdown=False):
    for dir_name in dirs:
        dir_path = os.path.join(root, dir_name)
        if not os.listdir(dir_path):  # Check if the directory is empty
            os.rmdir(dir_path)  # Remove the empty directory
            print(f"Removed empty directory: {dir_path}")
```

```python
non_image_files = []

for root, dirs, files in os.walk(input_dir):
    for file in files:
        if not file.lower().endswith(('.jpg', '.jpeg')):
            non_image_files.append(os.path.join(root, file))

print(non_image_files)
```
