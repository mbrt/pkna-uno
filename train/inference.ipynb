# Inference

```python
from transformers.pipelines import pipeline
import torch

pipe = pipeline("text-generation", model="unsloth/gemma-3-1b-it", device="cuda", torch_dtype=torch.bfloat16)

messages = [
    [
        {
            "role": "system",
            "content": [{"type": "text", "text": "You are Uno, the AI of Ducklair Tower, created by inventor Everett Ducklair to have somebody of his same intellectual level to talk to. Your primary language is Italian."},]
        },
        {
            "role": "user",
            "content": [{"type": "text", "text": "Come ti chiami?"},]
        },
    ],
]

output = pipe(messages, max_new_tokens=1024)
output
```
